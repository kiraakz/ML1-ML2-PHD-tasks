```python
#11111111111111111111111111111111
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
"""
numpy (np) — работает с числами и массивами.
matplotlib.pyplot (plt) — строит графики.
Ridge — вид линейной регрессии, которая не переобучается.
PolynomialFeatures — добавляет степени (x², x³, …).
make_pipeline — объединяет шаги (создание признаков + модель).
train_test_split — делит данные на обучение и тест.
mean_squared_error, r2_score — оценивают качество модели.
"""



# Определяем новую функци
def f(x):
    return x * np.sin(x**2)
"""
Функция, Это настоящая зависимость между x и y,
которую мы будем пытаться приблизить моделью.

"""


# Создаем данные
x = np.linspace(0, 5, 50)
y = f(x)
"""
создаёт 50 равномерных чисел от 0 до 5.
Затем мы вычисляем y для каждого x.
Получаются 50 точек на графике.
"""


# Делим на train и test
X = x[:, np.newaxis]
"""
Преобразуем x в двумерный массив (так требует scikit-learn).
Теперь X имеет форму (50, 1) — 50 строк, 1 колонка.
"""


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
"""
train — на них учится модель
test — на них проверяется, насколько хорошо модель предсказывает
random_state=42 — просто фиксирует случайность (чтобы результат был одинаковым при повторном запуске).
"""


# 4. Подбор степени регрессии
best_degree = None
for degree in range(1, 10): # Проверяем степени от 1 до 9. (1 — почти прямая линия, 9 — сильно изогнутая кривая).
    model = make_pipeline(PolynomialFeatures(degree), Ridge())
    """
    Создаём “конвейер”:
    PolynomialFeatures(degree) — добавляет степени x, x², x³, …
    Ridge() — обучает модель, не позволяя ей переобучиться (регуляризация).
    """
    model.fit(X_train, y_train) #Обучаем модель на тренировочных данных.
    y_pred = model.predict(X_test) #Предсказываем значения на тестовых данных.
    mse = mean_squared_error(y_test, y_pred)
    print(f"Degree={degree}, MSE={mse:.2f}")
    if mse <= 10:
        best_degree = degree
        break
    """
    Если ошибка стала меньше или равна 10,
    мы запоминаем эту степень как “лучшую” и выходим из цикла.
    """

# 5. Строим график с лучшей степенью
model = make_pipeline(PolynomialFeatures(best_degree), Ridge())
model.fit(X_train, y_train)
y_plot = model.predict(X)
"""
Строим модель заново, но уже с лучшей найденной степенью
и предсказываем значения для всех X, чтобы нарисовать гладкую кривую.
"""

plt.figure(figsize=(8,5))
plt.scatter(X_train, y_train, color='black', marker='o', label='train points')
plt.scatter(X_test, y_test, color='red', marker='x', label='test points')
plt.plot(X, y_plot, color='teal', label=f'Polynomial degree={best_degree}')
plt.legend()
plt.show()

"""
plt.scatter() — точки данных:
черные кружки (train)
красные крестики (test)
plt.plot() — линия предсказаний модели.
plt.legend() — подписи к элементам.
plt.title() — название графика.
plt.show() — показать график.

На графике видно, что большинство точек расположены близко к линии регрессии,
что свидетельствует о хорошем приближении данных моделью.
Это значит, что предсказанные значения y_pred мало отличаются от реальных значений y.
Следовательно, модель имеет низкую ошибку (MSE) и хорошо описывает зависимость между переменными.
"""
```

```python
#222222222222222222222222222222222222222222222222222222

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

"""
numpy — для работы с числами и массивами
matplotlib.pyplot — для построения графиков
PolynomialFeatures — создаёт новые признаки для полиномиальной регрессии
LinearRegression — сама модель линейной регрессии
mean_squared_error, r2_score — метрики для оценки точности
"""

np.random.seed(0) #фиксируем генератор случайных чисел (чтобы результат не менялся при каждом запуске)
X = np.linspace(0, 10, 30) #создаём 30 точек X от 0 до 10
y = 0.5 * X**2 - 2*X + 3 + np.random.randn(30) * 3  
"""
формула зависимости y от x, к которой добавлен случайный шум (np.random.randn(30) * 3), имитирующий реальные погрешности
"""
X = X[:, np.newaxis] #превращаем одномерный массив X в двумерный

plt.figure(figsize=(10, 6)) #создаём окно для графика (10×6 дюймов)
colors = ['red', 'green', 'blue', 'orange']  # цвета для кривых
degrees = [2, 3, 7, 9] #степени полинома, которые будем проверять

for count, degree in enumerate(degrees):
    poly = PolynomialFeatures(degree=degree) 
    """
    добавляет новые столбцы в X:
    например, если degree=3, то будет [1, x, x², x³]
    """
    X_poly = poly.fit_transform(X) #применяем это преобразование к нашим данным

    model = LinearRegression()
    model.fit(X_poly, y)
    """
    создаём объект линейной регрессии
    обучаем её на новых данных (X_poly, y)
    """
    y_pred = model.predict(X_poly) #модель вычисляет предсказанные y для каждого x

    mse = mean_squared_error(y, y_pred) #среднеквадратичная ошибка, чем меньше, тем лучше
    r2 = r2_score(y, y_pred) 
    """
    коэффициент детерминации (R²), показывает насколько модель объясняет данные
    (R² близко к 1 — модель отличная, R² ≈ 0 — слабая)
    """
    print(f"degree={degree}")
    print(f"MSE={mse:.2f}")
    print(f"variance (R²)={r2:.2f}")
    print()
    plt.plot(X, y_pred, color=colors[count], label=f'degree={degree}, MSE={mse:.2f}')

plt.scatter(X, y, color='black', label='Исходные точки')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

"""
Степень 2 (красная линия):
Почти идеально совпадает с формой исходных данных.
Большинство точек расположены близко к линии, что говорит о хорошем приближении и малой ошибке (MSE).

Степень 3 (зелёная линия):
Также даёт хорошее приближение, но начинает слегка изгибаться в областях, где данных меньше.
Ошибка (MSE) немного выше, но всё ещё низкая.
Модель всё ещё описывает зависимость адекватно, но чуть чувствительнее к шуму.

Степень 7 (синяя линия):
Кривая начинает “колебаться” между точками.
Некоторые участки линии проходят точно через точки, но другие отклоняются.
Это говорит о начале переобучения — модель слишком подстраивается под шум в данных.

Степень 9 (оранжевая линия):
Сильно извивается, особенно на краях графика.
Модель пытается точно пройти через каждую точку, но теряет общую тенденцию.
Ошибка (MSE) может быть маленькой на обучающих данных, но такая модель плохо обобщает данные — это переобучение.
"""
```

```python
#3333333333333333333333333333333


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split

# ---------- 0. Параметры ----------
csv_path = 'DeepLearning_v.1.0.csv'   # <- укажи путь к файлу, если в другой папке
search_substring = 'convolutional'   # ищем строку с этим текстом (без учета регистра)
max_degree = 12                       # максимум степени, который проверим

# ---------- 1. Загрузка данных ----------
sc_df = pd.read_csv(csv_path, encoding='latin1')
print('Columns (headers):', list(sc_df.columns))

# ---------- 2. Находим нужную строку ----------
# Попробуем найти строку по подстроке 'convolutional' в столбце 'Field'
mask = sc_df['Field'].str.lower().str.contains(search_substring)
if mask.sum() == 0:
    # если не нашли - можно взять n=1 как в описании
    print("Подстрока не найдена — возьмём строку n=1 (вторая строка файла).")
    n = 1
    row = sc_df.loc[n]
else:
    # берём первую найденную строку
    idx = sc_df[mask].index[0]
    print(f"Найдена строка с индексом {idx}:")
    row = sc_df.loc[idx]

print('Field label:', row['Field'])

# ---------- 3. Извлекаем годы (колонки) и значения y ----------
# Предположим, что все оставшиеся колонки после 'Field' — это годы (или числовые метки).
years_cols = sc_df.columns[1:]   # все колонки, кроме 'Field'
# Попробуем конвертировать названия колонок в int (вдруг есть лишние колонки — их пропустим)
years = []
for col in years_cols:
    try:
        years.append(int(col))
    except:
        # если колонка не парсится в int — пропускаем
        pass

years = np.array(years)
print('Годы (используем):', years)

# Собираем y — значения из строки для тех годов, которые удалось распарсить
# Берём колонки как строки, чтобы индекс совпал
y_list = []
for col in years_cols:
    try:
        y_list.append(float(row[col]))
    except:
        # если не удалось (NaN или текст) — ставим np.nan
        y_list.append(np.nan)

y_arr = np.array(y_list, dtype=float)

# Оставим только элементы, где год был распознан (years) и значение не NaN
# Для этого сделаем маску по длине years (они идут в порядке колонок), но безопаснее:
valid_mask = ~np.isnan(y_arr)
# Иногда колонок может быть больше — отрежем по длине years
valid_mask = valid_mask[:len(years)]
x_years = years[valid_mask]
y = y_arr[:len(years)][valid_mask]

print('Используемые пары (год, value):')
for a,b in zip(x_years, y):
    print(a, int(b))

# ---------- 4. Нормировка/сдвиг X (очень важно) ----------
# Для численных стабильных вычислений лучше сместить годы ближе к нулю.
# Часто удобно: x = year - 2000 (или -2010), в зависимости от годов.
x = x_years - 2000
X = x.reshape(-1, 1)

# ---------- 5. Перебор степеней и поиск минимальной степени для R2 > 0.99 ----------
best_degree_train = None   # условие на R2 на всех данных (train-fit)
best_degree_test = None    # условие на R2 на test (train/test split)
results = []

# подготовим train/test split для проверки обобщения
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

for degree in range(1, max_degree + 1):
    # Создаём полиномиальные признаки и обучаем модель
    poly = PolynomialFeatures(degree=degree, include_bias=True)
    X_poly = poly.fit_transform(X)           # для оценки на всех данных
    Xtr_poly = poly.transform(X_train)       # для трейн/тест
    Xte_poly = poly.transform(X_test)
    
    model = LinearRegression()
    model.fit(Xtr_poly, y_train)             # обучаем на train (чтобы test R2 имел смысл)
    
    # предсказания
    y_pred_all = model.predict(X_poly)       # предсказание на всех точках
    y_pred_test = model.predict(Xte_poly)    # предсказание на тестовой выборке
    
    r2_all = r2_score(y, y_pred_all)         # R2 на всех данных
    r2_test = r2_score(y_test, y_pred_test)  # R2 на test (объективная оценка)
    mse_all = mean_squared_error(y, y_pred_all)
    mse_test = mean_squared_error(y_test, y_pred_test)
    
    results.append((degree, r2_all, r2_test, mse_all, mse_test))
    
    print(f"degree={degree:2d} | R2_all={r2_all:.4f} | R2_test={r2_test:.4f} | MSE_all={mse_all:.2f} | MSE_test={mse_test:.2f}")
    
    # отмечаем минимальную степень, где R2 на всех данных > 0.99 (если ещё не отмечена)
    if (best_degree_train is None) and (r2_all > 0.99):
        best_degree_train = degree
    # отмечаем минимальную степень, где R2 на test > 0.99
    if (best_degree_test is None) and (r2_test > 0.99):
        best_degree_test = degree

print()
print('Минимальная степень, где R2 на всех данных > 0.99:', best_degree_train)
print('Минимальная степень, где R2 на test (train/test split) > 0.99:', best_degree_test)

# ---------- 6. Построим график: точки + кривые для нескольких степеней (до лучшей / нескольких) ----------
plt.figure(figsize=(10,6))
plt.scatter(X.flatten(), y, color='black', label='Данные (x=year-2000)')

# нарисуем кривые для всех проверенных степеней (или ограничим до 8 для читаемости)
plot_degrees = [d for d, *_ in results if d <= min(max_degree, 10)]
colors = plt.cm.viridis(np.linspace(0,1,len(plot_degrees)))

x_plot = np.linspace(X.min(), X.max(), 200).reshape(-1,1)  # плотная сетка для гладкой линии

for color, degree in zip(colors, plot_degrees):
    poly = PolynomialFeatures(degree=degree, include_bias=True)
    Xp = poly.fit_transform(x_plot)
    # переобучим модель на всех данных для красивой линии (так линии согласованы с тем, что показывал r2_all)
    model = LinearRegression()
    model.fit(poly.fit_transform(X), y)
    y_plot = model.predict(Xp)
    plt.plot(x_plot, y_plot, color=color, label=f'degree={degree}')

plt.xlabel('year - 2000')
plt.ylabel('publications count')
plt.title('Polynomial fits for different degrees (convolutional row)')
plt.legend()
plt.grid(True)
plt.show()

```
