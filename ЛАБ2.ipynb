```python
#11111111111111111111111111111111
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
"""
numpy (np)      .
matplotlib.pyplot (plt)   .
Ridge    ,   .
PolynomialFeatures    (x, x, ).
make_pipeline    (  + ).
train_test_split       .
mean_squared_error, r2_score    .
"""



#   
def f(x):
    return x * np.sin(x**2)
"""
,     x  y,
     .

"""


#  
x = np.linspace(0, 5, 50)
y = f(x)
"""
 50    0  5.
   y   x.
 50   .
"""


#   train  test
X = x[:, np.newaxis]
"""
 x    (  scikit-learn).
 X   (50, 1)  50 , 1 .
"""


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
"""
train     
test    ,    
random_state=42     (      ).
"""


# 4.   
best_degree = None
for degree in range(1, 10): #    1  9. (1    , 9    ).
    model = make_pipeline(PolynomialFeatures(degree), Ridge())
    """
     :
    PolynomialFeatures(degree)    x, x, x, 
    Ridge()   ,     ().
    """
    model.fit(X_train, y_train) #    .
    y_pred = model.predict(X_test) #    .
    mse = mean_squared_error(y_test, y_pred)
    print(f"Degree={degree}, MSE={mse:.2f}")
    if mse <= 10:
        best_degree = degree
        break
    """
          10,
             .
    """

# 5.     
model = make_pipeline(PolynomialFeatures(best_degree), Ridge())
model.fit(X_train, y_train)
y_plot = model.predict(X)
"""
  ,      
     X,    .
"""

plt.figure(figsize=(8,5))
plt.scatter(X_train, y_train, color='black', marker='o', label='train points')
plt.scatter(X_test, y_test, color='red', marker='x', label='test points')
plt.plot(X, y_plot, color='teal', label=f'Polynomial degree={best_degree}')
plt.legend()
plt.show()

"""
plt.scatter()   :
  (train)
  (test)
plt.plot()    .
plt.legend()    .
plt.title()   .
plt.show()   .

  ,        ,
      .
 ,    y_pred      y.
,     (MSE)      .
"""
```

```python
#222222222222222222222222222222222222222222222222222222

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

"""
numpy       
matplotlib.pyplot    
PolynomialFeatures       
LinearRegression     
mean_squared_error, r2_score     
"""

np.random.seed(0) #    (      )
X = np.linspace(0, 10, 30) # 30  X  0  10
y = 0.5 * X**2 - 2*X + 3 + np.random.randn(30) * 3  
"""
  y  x,      (np.random.randn(30) * 3),   
"""
X = X[:, np.newaxis] #   X  

plt.figure(figsize=(10, 6)) #    (106 )
colors = ['red', 'green', 'blue', 'orange']  #   
degrees = [2, 3, 7, 9] # ,   

for count, degree in enumerate(degrees):
    poly = PolynomialFeatures(degree=degree) 
    """
        X:
    ,  degree=3,   [1, x, x, x]
    """
    X_poly = poly.fit_transform(X) #     

    model = LinearRegression()
    model.fit(X_poly, y)
    """
       
         (X_poly, y)
    """
    y_pred = model.predict(X_poly) #   y   x

    mse = mean_squared_error(y, y_pred) # ,  ,  
    r2 = r2_score(y, y_pred) 
    """
      (R),     
    (R   1   , R  0  )
    """
    print(f"degree={degree}")
    print(f"MSE={mse:.2f}")
    print(f"variance (R)={r2:.2f}")
    print()
    plt.plot(X, y_pred, color=colors[count], label=f'degree={degree}, MSE={mse:.2f}')

plt.scatter(X, y, color='black', label=' ')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

"""
 2 ( ):
      .
     ,         (MSE).

 3 ( ):
   ,      ,   .
 (MSE)  ,    .
     ,     .

 7 ( ):
    .
      ,   .
            .

 9 ( ):
 ,    .
      ,    .
 (MSE)      ,         .
"""
```

```python
#3333333333333333333333333333333


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split

# ---------- 0.  ----------
csv_path = 'DeepLearning_v.1.0.csv'   # <-    ,    
search_substring = 'convolutional'   #      (  )
max_degree = 12                       #  ,  

# ---------- 1.   ----------
sc_df = pd.read_csv(csv_path, encoding='latin1')
print('Columns (headers):', list(sc_df.columns))

# ---------- 2.    ----------
#      'convolutional'   'Field'
mask = sc_df['Field'].str.lower().str.contains(search_substring)
if mask.sum() == 0:
    #    -   n=1   
    print("      n=1 (  ).")
    n = 1
    row = sc_df.loc[n]
else:
    #    
    idx = sc_df[mask].index[0]
    print(f"    {idx}:")
    row = sc_df.loc[idx]

print('Field label:', row['Field'])

# ---------- 3.   ()   y ----------
# ,      'Field'    (  ).
years_cols = sc_df.columns[1:]   #  ,  'Field'
#      int (      )
years = []
for col in years_cols:
    try:
        years.append(int(col))
    except:
        #      int  
        pass

years = np.array(years)
print(' ():', years)

#  y       ,   
#    ,   
y_list = []
for col in years_cols:
    try:
        y_list.append(float(row[col]))
    except:
        #    (NaN  )   np.nan
        y_list.append(np.nan)

y_arr = np.array(y_list, dtype=float)

#   ,     (years)    NaN
#       years (    ),  :
valid_mask = ~np.isnan(y_arr)
#          years
valid_mask = valid_mask[:len(years)]
x_years = years[valid_mask]
y = y_arr[:len(years)][valid_mask]

print('  (, value):')
for a,b in zip(x_years, y):
    print(a, int(b))

# ---------- 4. / X ( ) ----------
#          .
#  : x = year - 2000 ( -2010),    .
x = x_years - 2000
X = x.reshape(-1, 1)

# ---------- 5.        R2 > 0.99 ----------
best_degree_train = None   #   R2    (train-fit)
best_degree_test = None    #   R2  test (train/test split)
results = []

#  train/test split   
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

for degree in range(1, max_degree + 1):
    #      
    poly = PolynomialFeatures(degree=degree, include_bias=True)
    X_poly = poly.fit_transform(X)           #     
    Xtr_poly = poly.transform(X_train)       #  /
    Xte_poly = poly.transform(X_test)
    
    model = LinearRegression()
    model.fit(Xtr_poly, y_train)             #   train ( test R2  )
    
    # 
    y_pred_all = model.predict(X_poly)       #    
    y_pred_test = model.predict(Xte_poly)    #    
    
    r2_all = r2_score(y, y_pred_all)         # R2   
    r2_test = r2_score(y_test, y_pred_test)  # R2  test ( )
    mse_all = mean_squared_error(y, y_pred_all)
    mse_test = mean_squared_error(y_test, y_pred_test)
    
    results.append((degree, r2_all, r2_test, mse_all, mse_test))
    
    print(f"degree={degree:2d} | R2_all={r2_all:.4f} | R2_test={r2_test:.4f} | MSE_all={mse_all:.2f} | MSE_test={mse_test:.2f}")
    
    #   ,  R2    > 0.99 (   )
    if (best_degree_train is None) and (r2_all > 0.99):
        best_degree_train = degree
    #   ,  R2  test > 0.99
    if (best_degree_test is None) and (r2_test > 0.99):
        best_degree_test = degree

print()
print(' ,  R2    > 0.99:', best_degree_train)
print(' ,  R2  test (train/test split) > 0.99:', best_degree_test)

# ---------- 6.  :  +     (  / ) ----------
plt.figure(figsize=(10,6))
plt.scatter(X.flatten(), y, color='black', label=' (x=year-2000)')

#       (   8  )
plot_degrees = [d for d, *_ in results if d <= min(max_degree, 10)]
colors = plt.cm.viridis(np.linspace(0,1,len(plot_degrees)))

x_plot = np.linspace(X.min(), X.max(), 200).reshape(-1,1)  #     

for color, degree in zip(colors, plot_degrees):
    poly = PolynomialFeatures(degree=degree, include_bias=True)
    Xp = poly.fit_transform(x_plot)
    #         (    ,   r2_all)
    model = LinearRegression()
    model.fit(poly.fit_transform(X), y)
    y_plot = model.predict(Xp)
    plt.plot(x_plot, y_plot, color=color, label=f'degree={degree}')

plt.xlabel('year - 2000')
plt.ylabel('publications count')
plt.title('Polynomial fits for different degrees (convolutional row)')
plt.legend()
plt.grid(True)
plt.show()

```
